#상속

프로그래밍을 하다 보면 각 class에서 비슷한 변수나 함수를 중복해서 사용하는 경우가 있다.

또한 현재의 클래스를 변형하지 않는 상태에서 다른 기능을 추가해야 하는 경우도 있다.

```python
class Goods:
    name = '상품명'
    price = 0

class Sale_goods(Goods):
    def make(self, name, price):
        self.name = name
        self.price = price

    def print_goods(self):
        print(self.name)
        print(self.price)

sale_cola = Sale_goods()
sale_cola.print_goods()
print('*'*20)
sale_cola.make('콜라', 1500)
sale_cola.print_goods()
```

Sale_goods 클래스는 Goods 클래스를 상속받았다. 그래서 Goods 클래스를 '부모 클래스', Sale_goods를 '자식 클래스'라고 한다. sale_cola 객체는 Sale_goods 클래스의 인스턴스로, sale_cola 객체의 print_goods함수를 실행했을 때 상속받은 name과 price 변숫값을 출력하는 것을 확인할 수 있다. sale_cola객체의 make 함수에 '콜라', 1500 인수를 넣어 실행한 결과, 실제 Sale_goods 클래스에는 name과 price가 없지만, Goods 클래스를 상속받았기 때문에 name과 price 변수에 값이 입력되고 print_goods 함수 실행 시 바뀐 값으로 출력되는 것을 확인

## 상속과 생성자 관계

```python
class Goods:
    def __init__(self):
        print('Goods클래스 생성자')

class Sale_goods(Goods):
    pass

sale_cola = Sale_goods()
```

부모 클래스에만 생성자가 있을 경우 객체를 생성하였을 때, 부모 클래스의 생성자가 실행된다.

```python
class Goods:
    def __init__(self):
        print('Goods클래스 생성자')

class Sale_goods(Goods):
    def __init__(self):
        print('Sale_goods 클래스 생성자')
    pass

sale_cola = Sale_goods()
```

부모 클래스와 자식 클래스 모두 생성자가 있는 경우, 자식 클래스의 생성자만 실행되고 부모 클래스의 생성자는 무시된다.

```python
class Goods:
    def __init__(self, name):
        self.name = name
        print(self.name)
        print("Goods 클래스 생성자")

class Sale_goods(Goods):
    def __init__(self, name):
        super().__init__(name)
        print('Sale_goods 클래스 생성자')
    pass

sale_cola = Sale_goods('콜라')
```

부모 클래스의 생성자를 호출해야 할 경우, 자식 클래스에서 super()를 사용하여 부모 클래스의 생성자를 호출할 수 있다.

super()란?

super()는 자식클래스에서 상속받은 부모 클래스를 호출할 때 사용하는 함수이다.
